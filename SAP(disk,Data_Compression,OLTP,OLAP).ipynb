# **What is a Disk?**

**A disk (or hard disk) is a storage device in a computer where data is saved permanently. It is used to store files, programs, and the operating system, even when the computer is turned off.**

**Types of Disks:**

HDD (Hard Disk Drive) – Uses spinning magnetic disks to store data (slower).

SSD (Solid-State Drive) – Uses flash memory (faster and more reliable).

**Example:**

Think of a disk like a notebook where you write important notes. Even if you

close the notebook, the notes remain, just like data on a disk.

# **what is mean by Data Compression in SAP HANA?**

**Data compression in SAP HANA is a technique used to reduce the size of stored data in memory and on disk. It helps in saving storage space and improving performance by reducing the amount of data that needs to be processed.**

# **Benefits of Compression in SAP HANA:**

✅ Saves Storage Space – Uses less memory to store data.

✅ Faster Processing – Smaller data means faster access and processing.

✅ Better Performance – Improves query performance and system efficiency.

#**what is mean by OLTP (Online Transaction Processing) ?**

Used for day-to-day transactions (fast and frequent).

Handles many small transactions like inserting, updating, and deleting records.

Focuses on real-time processing and quick response time.

✅ **Example**:

Bank Transactions – When you withdraw money from an ATM, it updates your account balance instantly.

E-commerce – When you place an order on Amazon, the stock is updated immediately.

# **what is mean by OLAP (Online Analytical Processing)**

Used for data analysis and reporting (big and complex queries).

Handles large amounts of historical data.

Focuses on decision-making and generating reports.

✅ **Example:**

Sales Analysis – A company analyzes yearly sales trends to decide which products to stock.

Business Intelligence – A manager generates a report to see which region had the highest sales in the last 5 years.
