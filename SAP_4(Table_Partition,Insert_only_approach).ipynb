# **what is mean by Insert-Only Approach in SAP HANA ?**

**The Insert-Only Approach in SAP HANA means that data is only inserted and not modified or deleted directly. Instead of updating existing records, new records are added, and the system periodically merges them to optimize performance.**

**Main and Delta Storage in SAP HANA**

SAP HANA uses a two-layer storage system:

# **Explanation of the Image**
The image explains how data merging works in SAP HANA:

**Prior to Merge:**

New records are written to the Delta 1 storage.

Reads access both Main 1 and Delta 1 for data retrieval.

**During Merge:**

Data from Main 1 and Delta 1 are combined into a new Main 2.

A new Delta 2 storage starts collecting new inserts.

Read operations access both Main 1, Main 2, Delta 1, and Delta 2 temporarily.

**After Merge:**

Main 2 replaces Main 1 as the primary storage.

Delta 2 is now the active delta storage for new data.

Read operations now access Main 2 and Delta 2 only.

**Conclusion**

This Insert-Only + Merge Approach ensures better performance by avoiding frequent updates and deletions. Data remains structured and optimized for fast queries in SAP HANA.
